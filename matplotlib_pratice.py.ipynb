# Subplot


```python
#subplot (it takes 3 arrguments for complete their condition out of 3 1st takes row 2nd takes column 3rd takes plot num)



import matplotlib.pyplot as plt
import numpy as np


x=([12,32,43,54,56,43])
y=([23,45,46,57,78,90])


plt.subplot(1,2,1)  #1 row 2 column 2 plt.number
plt.plot(x,y)




x=([12,2,43,54,6,43])
y=([23,5,46,57,8,90])


plt.subplot(1,2,2)   #1 row 2 column 2 plt.number
plt.plot(x,y)
plt.show()
```

# Pie()

```python
#pie()
import matplotlib.pyplot as plt
import numpy as np

x=np.array([12,23,43,25,65])

plt.pie(x)
plt.show()
```

```python
#pie: labels
import matplotlib.pyplot as plt
import numpy as np

x=np.array([12,23,43,25,65])
r=np.array(["ram","sham","gham","kam","vam"])
plt.pie(x,labels=r,)
plt.show()
```

```python
#pie: colours
import matplotlib.pyplot as plt
import numpy as np

x=np.array([12,23,43,25,65])
r=np.array(["r","g","y","k","b"])
rs=np.array(["ram","sham","gham","kam","vam"])
plt.pie(x,labels=rs,colors=r)
plt.show()
```

```python
#pie: starangle ( bydefault the graph is stars anticlockwise from 3:00 but we  can change them )
import matplotlib.pyplot as plt
import numpy as np

x=np.array([12,23,43,25,65])
r=np.array(["r","g","y","k","b"])
rs=np.array(["ram","sham","gham","kam","vam"])
plt.pie(x,labels=rs,colors=r,startangle=90)
plt.show()
```

```python
#pie: Explode
import matplotlib.pyplot as plt
import numpy as np

x=np.array([12,23,43,25,65])
r=np.array(["r","g","y","k","b"])
rs=np.array(["ram","sham","gham","kam","vam"])
v=np.array([0.2,0.1,0.2,0.2,0.5])
plt.pie(x,labels=rs,colors=r,explode=v)
plt.show()
```

```python
#pie: Explode
import matplotlib.pyplot as plt
import numpy as np

x=np.array([12,23,43,25,65])
r=np.array(["r","g","y","k","b"])
rs=np.array(["ram","sham","gham","kam","vam"])
v=np.array([0.2,0.1,0.2,0.2,0.5])
plt.pie(x,labels=rs,colors=r,explode=v,shadow=True)
plt.show()
```

```python
#pie: legend :it show the labels inside the table
import matplotlib.pyplot as plt
import numpy as np

x=np.array([12,23,43,25,65])
r=np.array(["r","g","y","k","b"])
rs=np.array(["ram","sham","gham","kam","vam"])
v=np.array([0.2,0.1,0.2,0.2,0.5])
plt.pie(x,labels=rs,colors=r,explode=v)
plt.legend(title="legend")
plt.show()
```
